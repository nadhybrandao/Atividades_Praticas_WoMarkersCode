# Desafio Prático 1: Visualizando Dados Diferentes

**Descrição:** Neste desafio, suas alunas irão visualizar dados de diferentes formas utilizando Matplotlib, Seaborn e Plotly. Elas deverão criar gráficos que mostrem a distribuição dos dados, correlações entre variáveis e tendências.

**Tarefas:**
1. Carregar os dados de um arquivo CSV fornecido (saude.csv).
2. Criar um histograma da variável 'age' usando Matplotlib.
3. Criar um gráfico de dispersão mostrando a relação entre 'height' e 'weight' usando Seaborn.
4. Criar um gráfico interativo de barras mostrando a contagem de indivíduos por
'gender' usando Plotly.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('saude.csv')
print(data.head(10))
```

```python
# Histograma da variável 'age'
plt.figure(figsize=(5, 5))
plt.hist(data['age'], bins=5, color='purple', edgecolor='black')
plt.title('Distribuição da Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de dispersão entre 'height' e 'weight'
plt.figure(figsize=(5, 5))
sns.scatterplot(data=data, x='height', y='weight', hue='gender', palette='viridis')
plt.title('Relação entre Altura e Peso')
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.show()
```

```python
# Gráfico de barras interativo para contagem de 'gender'
fig, (ax1) = plt.subplots(1, figsize=(5, 5))

sns.countplot(data=data, x='gender', ax=ax1)
plt.title('Contagem por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Contagem')
plt.show()
```

# Desafio Prático 2: Séries Temporais
**Descrição:** Neste desafio, suas alunas irão analisar uma série temporal de dados de vendas. Elas deverão criar gráficos que mostrem as vendas ao longo do tempo, identificar tendências e sazonalidades.

**Tarefas:**
1. Carregar os dados de um arquivo CSV fornecido (vendas.csv).
2. Criar um gráfico de linhas das vendas ao longo do tempo usando Matplotlib.
3. Criar um gráfico de decomposição sazonal usando Seaborn.
4. Criar um gráfico interativo que permita a exploração dos dados de vendas ao longo do tempo usando Plotly

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import plotly.express as px
```

```python
# Carregar os dados e converter a coluna de datas para o tipo datetime
data = pd.read_csv('vendas.csv')
data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True) # Definindo a coluna de data como índice

print(data.head())
```

```python
print(plt.style.available)
```

```python
# Gráfico de linhas das vendas ao longo do tempo
plt.style.use('fast')
plt.figure(figsize=(15, 5))
plt.plot(data.index, data['sales'], color='blue')
plt.title('Vendas ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Vendas')
plt.show()
```

```python
# Decomposição sazonal
decomposition = seasonal_decompose(data['sales'], model='additive', period=12)

# Plotando a decomposição
plt.figure(figsize=(10, 8))
plt.subplot(4, 1, 1)
plt.plot(data['sales'], label='Vendas')
plt.title('Série Original')
plt.subplot(4, 1, 2)
plt.plot(decomposition.trend, label='Tendência')
plt.title('Tendência')
plt.subplot(4, 1, 3)
plt.plot(decomposition.seasonal, label='Sazonalidade')
plt.title('Sazonalidade')
plt.subplot(4, 1, 4)
plt.plot(decomposition.resid, label='Resíduo')
plt.title('Resíduo')
plt.tight_layout()
plt.show()
```

```python
# Gráfico interativo de vendas ao longo do tempo
fig = px.line(data, x=data.index, y='sales', title='Vendas ao Longo do Tempo')
fig.update_xaxes(title_text='Data')
fig.update_yaxes(title_text='Vendas')
fig.show()
```
