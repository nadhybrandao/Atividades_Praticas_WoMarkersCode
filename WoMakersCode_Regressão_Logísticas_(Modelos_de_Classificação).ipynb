````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
````

````python
# Carregar a base de dados
data = pd.read_csv('iris.csv')
print(data.head(10))
````

# Questão 1.A
Faça uma análise inicial sobre esse dataset

````python
# Carregar o dataset Iria
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Mapeando os valores das espécies
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
print(df.head())
````

# Questão 1.B
Use o boxplot e o histograma para caracterizar as propriedades de cada uma das espécies existentes.


````python
# Boxplot para cada variável por espécie
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.columns[:-1],1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='species', y=column, data=df)
    plt.title(f'Boxplot de {column}')

plt.tight_layout()
plt.show()

# Histograma
df.hist(by='species', figsize=(12, 8))
plt.show()
````

# Questão 1.D
Aplique a regressão logística para avaliar o modelo declassificação. Calcule a acurácia, precisão e recall.

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix

# Dividir dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

# Regressão Logística
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
````

# Questão 1.E
Calcule a acurácia, precisão e recall.

````python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')

print(f"Acurácia: {accuracy}")
print(f"Precisão: {precision}")
print(f"Recall: {recall}")
````

# Questão 1.F
Plote a matriz de confusão com matplotlib ou seaborn.

````python
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
````

````python
from sklearn.datasets import load_digits

# Carregar o dataset
digits = load_digits()
print(f"Quantidade de dados: {len(digits.data)}")
print(f"Quantidade de features: {digits.data.shape[1]}")
print(f"Verificação de dados nulos: {pd.DataFrame(digits.data).isnull().sum().sum()}")
print(f"Todos os dados são numéricos? {pd.DataFrame(digits.data).dtypes.all()}")
````

# Questão 2.B

Aplique a regressão logística para construir e avaliar o modelo de classificação.

````python
# Dividir dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.3, random_state=42)

# Modelo de Regressão Logística
model_digits = LogisticRegression(max_iter=200)
model_digits.fit(X_train, y_train)
y_pred_digits = model_digits.predict(X_test)
````

# Questão 2.C

Calcule a acurácia, precisão e recall com base no desempenho do modelo.

````python
# Métricas de avaliação
accuracy = accuracy_score(y_test, y_pred_digits)
precision = precision_score(y_test, y_pred_digits, average='weighted')
recall = recall_score(y_test, y_pred_digits, average='weighted')

print(f"Acurácia: {accuracy}")
print(f"Precisão: {precision}")
print(f"Recall: {recall}")
````

# Questão 2.D
Plote a matriz de confusão dos resultados do modelo utilizando matplotlib ou seaborn.

````python
conf_matrix = confusion_matrix(y_test, y_pred_digits)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
````
