## Desafio 2 ##

Definição do problema: Dada a tabela com a média de altura e peso de mulheres estadunidenses, onde os nomes das colunas estão em inglês height (altura) e weight (peso) e seus respectivos valores estão nas seguintes unidades: polegada e libras.

Construa uma nova representação onde os nomes das colunas estão em português, e os valores de altura são convertidos para centímetros e os valores de peso convertidos para quilograma.
Plote o scatterplot de altura x peso da nova representação. Código do dataset: women

Dicas:
* 1 polegada = 2,54 centímetros
* 1 libra = 0,45 quilos
* use o método .plot.scatter()


```python
# Carregando o dataset 'women'
data('women')
df_w = data('women')
df_w.head()
```

```python
# Convertendo altura para centímetros e peso para quilogramas
# Renomeando as colunas
df_w['altura (cm)'] = df_w['height'] * 2.54
df_w['peso (kg)'] = df_w['weight'] * 0.45
women_pt = df_w[['altura (cm)', 'peso (kg)']]
women_pt.head()
```

```python
# Plotando o scatterplot
women_pt.plot.scatter(x='altura (cm)', y='peso (kg)', title='Altura x Peso')
```

## Desafio 5 ##

Definição do problema: Você está trabalhando em um jornal e o
editor quer que você responda as seguintes questões:
1. Qual a empresa mais valiosa da lista da Forbes 2000?
2. Construa uma tabela com as TOP 10 empresas mais lucrativas da
lista.
3. Qual a média de valores de empresa das cinco categorias mais
valiosas?

Código do dataset: Forbes2000

Dicas:
* Método .max()
* Método nlargest()
* Método .groupby()

```python
# Carregando o dataset 'Forbes2000'
df_f = data('Forbes2000')
df_f.head(2000)
```

```python
# 1. Qual a empresa mais valiosa da lista da Forbes 2000?
name_marketvalue = df_f.loc[df_f['marketvalue'].idxmax(), 'name']
print(f'A empresa mais valiosa é: {name_marketvalue}')
```

```python
# 2. Construa uma tabela com as TOP 10 empresas mais lucrativas da lista.
top_10_profits = df_f.nlargest(10, 'profits')
print('Top 10 empresas mais lucrativas:')
print(top_10_profits[['name', 'profits']])
```

```python
# 3. Qual a média de valores de empresa das cinco categorias mais valiosas?
med_marketvalue = df_f.groupby('category')['marketvalue'].mean().nlargest(5)
print('Média de valores das cinco categorias mais valiosas:')
print(med_marketvalue)
```

# Desafio 3

Definição do problema:Dado que você está trabalhando com um grupo
de historiadores que estão analisando o naufrágio do Titanic, informe
aos historiadores as seguintes questões?
1. Quantas pessoas sobreviveram e quantas morreram?
2. Quantas pessoas eram da primeira, segunda e terceira classe?
3. Qual a porcentagem de sobreviventes na primeira classe?
4. Qual a porcentagem de mortos da terceira classe?

Código do dataset: titanic

Dicas:
- value_counts() (Itens 1 e 2)
- seleção de subconjuntos (Itens 3 e 4)

```python
# Carregando o dataset 'titanic'
df_t = data('titanic')
df_t.head()
```

```python
# Quantas pessoas sobreviveram e quantas morreram?
survived = df_t['survived'].value_counts()['yes']
no_survived = df_t['survived'].value_counts()['no']
print(f'Número de sobreviventes: {survived}')
print(f'Número de mortos: {no_survived}')
```

```python
# Quantas pessoas eram da primeira, segunda e terceira classe?
qtd_class = df_t['class'].value_counts()
print('Quantidade de pessoas por classe:')
print(qtd_class)
```

```python
# Qual a porcentagem de sobreviventes na primeira classe?
## Cálculo levado em consideração percentual de sobreviventes apenas na primeira classe
porcentagem_sobreviventes_primeira_classe = (df_t.loc[df_t['class'] == '1st class', 'survived'].map({'yes': 1, 'no': 0}).mean() * 100)
print(f'Porcentagem de sobreviventes na primeira classe: {porcentagem_sobreviventes_primeira_classe:.2f}%')
```

```python
terceira_classe = df_t[df_t['class'] == '3rd class']

# Calculando a porcentagem de mortos da terceira classe em relação ao total de tripulantes
total_tripulantes = len(df_t)
mortos_terceira_classe = terceira_classe['survived'].value_counts()[0]
porcentagem_mortos_terceira_classe = (mortos_terceira_classe / total_tripulantes) * 100

print(f'Porcentagem de mortos da terceira classe em relação ao total de tripulantes: {porcentagem_mortos_terceira_classe:.2f}%')
```
