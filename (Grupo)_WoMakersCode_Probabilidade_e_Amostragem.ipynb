```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Carregando o arquivo JSON
df = pd.read_csv('./desafio-01-consumidor.csv')
df.head(10)
```

```python
# Criar dois DataFrames, um para pré e outro para pós-pandemia
pre_pandemia = df[df['pre-pandemia'] == 1]
pos_pandemia = df[df['pre-pandemia'] == 0]
```

```python
# Plotar histograma da variável idade
plt.figure(figsize=(10, 5))
sns.histplot(pre_pandemia['idade'], color='blue', kde=True, label='pre-pandemia')
sns.histplot(pos_pandemia['idade'], color='orange', kde=True, label='Pós-pandemia')
plt.title('Distribuição de Idade')
plt.legend()
plt.show()

# Plotar histograma da variável renda
plt.figure(figsize=(10, 5))
sns.histplot(pre_pandemia['renda'], color='blue', kde=True, label='Pré-pandemia')
sns.histplot(pos_pandemia['renda'], color='orange', kde=True, label='Pós-pandemia')
plt.title('Distribuição de Renda')
plt.legend()
plt.show()
```
